```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_csv('data.csv')
```

```python
#filtering
df = df.loc[df['Make'] == 'BMW']

#value counts for transm type
totals = df['Transmission Type'].value_counts()
#pie chart plot
ax = totals.plot.pie(
 # Largest segment at the top.
counterclock=False, # Draw segments clockwise.
autopct='%1.1f%%',

)
ax.set_title('''Transmission type pie chart''')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#selecting specific rows to plot scatter graph
mydata = df[["Engine HP", "MSRP"]].dropna(how="any")
# Now plot with matplotlib
vals = mydata.values
#ax = plt.scatter(vals[:, 0], vals[:, 1])
mydata.plot(kind='scatter', x='Engine HP', y='MSRP', title="Horse Power vs Price",figsize=(10,5))
```

```python
#create table
table = pd.crosstab(df['Vehicle Style'], df['Popularity'],
margins=True)
table = table.sort_values(by='All', ascending=False)
table.drop(index='All', inplace=True)
table
table.drop('All', axis='columns', inplace=True)
ax = table.plot.barh()
ax.invert_yaxis()
ax.set_title('Popularity for different vehicle styles')
ax.set_xlabel('Frequency')
ax.set_ylabel('Vehicle Styles')
plt.show()
```

```python
#group and sort
df5 = df.groupby(['Model']).mean()
df5 = df5.sort_values(by='MSRP', ascending = False)


ax3 = df5.plot.bar(y='MSRP', align='center', figsize=(10,5))
ax3.set_title('Prices for Different Models')
```

```python


totals = df['Engine Fuel Type'].value_counts()

ax = totals.plot.pie(
 # Largest segment at the top.
counterclock=False, # Draw segments clockwise.
startangle=0, 
autopct='%1.1f%%',
#explode = explode
)
ax.set_title('Engine fuel type')

#plt.legend(title="Legend", bbox_to_anchor=(1.0, 0.5))
#explode = (0, 0.1, 0, 0)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
table = pd.crosstab(df['Model'], df['Transmission Type'],
margins=True)
table = table.sort_values(by='All', ascending=False)
table.drop(index='All', inplace=True)
table
table.drop('All', axis='columns', inplace=True)
ax = table.plot.barh(figsize=(10,15),colormap='Dark2')
ax.invert_yaxis()
ax.set_title('Vehicle Models and their Transmission Types')
ax.set_xlabel('Frequency')
ax.set_ylabel('Vehicle Models')
plt.savefig('vehmodel.png')
plt.show()
```
